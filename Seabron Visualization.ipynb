# Distribution Plot

```python
vis = sns.distplot(df.InternetUsers, bins=20, color='r', hist=True, kde= True, vertical=False) # distplot know as distrinution plot
                                     # Plot the graph using One varaible kwon as 'UNIVARIATE ANALYSIS' 
```

# Box Plot

```python
vis_1 = sns.boxplot(data = df, x="IncomeGroup", y="BirthRate",color='b', 
                    hue='IncomeGroup', linewidth=3, dodge=True,)  # Plot the graph using Two varaible kwon as 'BI-VARIATE ANALYSIS'
```

# Linear Model Plot (lmplot)

```python
vis_2 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate',fit_reg=True, 
                   hue='IncomeGroup', height=10, aspect=1, x_bins=None, 
                   scatter=True, truncate=True, legend_out=False)
```
